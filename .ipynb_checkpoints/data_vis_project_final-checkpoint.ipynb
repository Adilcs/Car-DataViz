```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

Cars = pd.read_csv('C:/Users/Adil/data_vis_project_final/data.csv')
Cars.isna().sum()
```

# Cleaning the dataset

### The dataset had required a lot of cleaning, you will see me doing more cleaning for the individual graphs

The main cleaning I did was getting rid of the null values for the columns with low missing values, but for the 
Market Category values, there were too many missing so I dropped this column.

```python
CarsClean = Cars[Cars['Engine Fuel Type'].notna()]
CarsClean = CarsClean[CarsClean['Engine HP'].notna()]
CarsClean = CarsClean[CarsClean['Engine Cylinders'].notna()]
CarsClean = CarsClean[CarsClean['Number of Doors'].notna()]
CarsClean.drop('Market Category', inplace=True, axis=1)
```

# Graph 1

### In this graph I made correlations between the Horsepower of Vehicles and the price of Vehicles.

This correlations is very obvious, cars tend to be at a higher price when they have more horsepower.
I added the different drive types in the graph which really splits the data.
Most cheaper cars are front wheel drive because they are easier to produce and safer, also most entry cars have low horsepower.
The more expensive a car is, the less chance that it is front wheel drive.
Mid priced vehicles are All wheel drive, because the drive train is a little more expensive to make
When you move towards higher horsepower and price you see a lot of rear wheel drive and four wheel drive
Four wheel drive vehicles are usually pick up trucks and large suvs, which need a lot of horsepower to move.
Also the rear wheel drives tend to be higher in price because they tend to be performance/sport driven vehicles.

```python
CarsCleaned = CarsClean[CarsClean['MSRP'] < 80000]  
CarsCleaned = CarsCleaned[CarsCleaned['Year'] > 2015]  
CarsCleaned = CarsCleaned[CarsCleaned['Engine Cylinders'] > 0] 
```

```python
CarHP = sns.relplot(data = CarsCleaned, x="Engine HP",y = "MSRP", hue = "Driven_Wheels", alpha = .6).set(title='Car Price based on Power')
CarHP
```

# Graph 2

In this graph I compared how efficient cars were based on the amount of power they outputted. They turned out to have a negative correlation. I also found out that the drive type is also associated to the efficiency. Four wheel drive vehicles look to be the least efficient vehicles, rear wheel drive cars are the runner up.

```python
CarGas = sns.relplot(data = CarsCleaned, x="Engine HP",y = "city mpg", hue = "Driven_Wheels", alpha = .6).set(title='Car Efficiency based on Power')
CarGas
```

# Graph 3

In this graph, I looked to see how popular vehicles were by their type. I found out that Pick ups and SUVs together are more popular than sedans. This is a lot different than how it was 20 years ago. I read an article that stated that the US sells way more SUVs than sedans, which is why companies are starting to stop making sedans and focusing on their larger vehicles.

```python
Carsdf = CarsClean
Carsdf["Vehicle Style"].unique()
def style(row):
    if (row['Vehicle Style'] == "4dr SUV" or row['Vehicle Style'] == "2dr SUV" or row['Vehicle Style'] == "Convertible SUV")  :
        val = "SUV"
    elif (row['Vehicle Style'] == "Crew Cab Pickup" or row['Vehicle Style'] == "Regular Cab Pickup" 
          or row['Vehicle Style'] == "Extended Cab Pickup"):
        val = "Pickup"
    elif (row['Vehicle Style'] == "Passenger Minivan" or row['Vehicle Style'] == "Cargo Minivan" 
          or row['Vehicle Style'] == "Cargo Van" or row['Vehicle Style'] == "Passenger Van"):
        val = "Van"
    elif (row['Vehicle Style'] == "4dr Hatchback" or row['Vehicle Style'] == "2dr Hatchback"):
        val = "Hatchback"
    else:
        val = row['Vehicle Style']
    return val

Carsdf['Vehicle_Type_Simple'] = Carsdf.apply(style, axis=1)

Carsdf2 = Carsdf.groupby("Vehicle_Type_Simple")["Popularity"].sum().reset_index()
plt.figure(figsize=(16,4))
Carsdfplot = sns.barplot(data = Carsdf2, x="Vehicle_Type_Simple",y = "Popularity", alpha = .8).set(title='Car Popularity based on Type')
```

# Final Graph

I made a fun tool to see what type of cars some manufacturers make. I used dash to incorporate this so you will have to open it
in a new tab. This is by far my most interactive graph. You can filter by Price, Origin, Drive type, and Vehicle Type. 

Some correlations I've found :

1) If you narrow the search to just Pickup trucks, you will see that American companies make most of them, with Toyota being
the only other competitor for this field.
2) If you raise the MSRP, you will see that only luxury and high performance brands show up.
3) Hatchbacks are not very common within America, but this is the opposite compared to every other origin.

```python
CarsCleaned2 = CarsClean[CarsClean['MSRP'] < 1000000]  
def group(row):
    if (row['Make'] == "BMW" or row['Make'] == "Audi" or row['Make'] == "Mercedes-Benz" or row['Make'] == "Porsche" or row['Make'] == "Volkswagen" or row['Make'] == "Maybach")  :
        val = "German"
    elif (row['Make'] == "Chrysler"  or row['Make'] == "Plymouth"  or
            row['Make'] == "Pontiac" or row['Make'] == "GMC"  or row['Make'] == "Oldsmobile" or row['Make'] == "Ford"  or
            row['Make'] == "Cadillac" or row['Make'] == "Chevrolet" or row['Make'] == "Dodge" or row['Make'] == "Lincoln"  or
            row['Make'] == "Buick"  or row['Make'] == "HUMMER" ):
        val = "American"
    elif (row['Make'] == "FIAT"  or row['Make'] == "Ferrari" or row['Make'] == "Alfa Romeo" or row['Make'] == "Lamborghini" or
          row['Make'] == "Maserati"):
        val = "Italian"
    elif (row['Make'] == "Nissan" or row['Make'] == "Mazda" or row['Make'] == "Mitsubishi" or row['Make'] == "Toyota" or
          row['Make'] == "Honda" or row['Make'] == "Suzuki" or row['Make'] == "Subaru" or row['Make'] == "Acura" or
          row['Make'] == "Lexus" or row['Make'] == "Infiniti" or row['Make'] == "Scion"):
        val = "Japanese"
    elif (row['Make'] == "Hyundai" or row['Make'] == "Kia" or row['Make'] == "Genesis"):
        val = "Korean"
    elif (row['Make'] == "McLaren" or row['Make'] == "Bentley" or row['Make'] == "Rolls-Royce" or row['Make'] == "Aston Martin"
         or row['Make'] == "Land Rover" or row['Make'] == "Lotus"):
        val = "British"
    else:
        val = "Other"
    return val

CarsCleaned2['Origin'] = CarsCleaned2.apply(group, axis=1)
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
app.layout = html.Div([
    html.Label('Origin'),
    dcc.Checklist(
    options=[
        {'label': 'German', 'value': 'German'},
        {'label': 'American', 'value': 'American'},
        {'label': 'Japanese', 'value': 'Japanese'},
        {'label': 'Korean', 'value': 'Korean'},
        {'label': 'British', 'value': 'British'},
        {'label': 'Other', 'value': 'Other'}
    ],
    value=['American'],
    id='checklist',
    labelStyle={'display': 'inline-block'}
    ),
    html.Label('Drive type'),
    dcc.Checklist(
    options=[
        {'label': 'Front Wheel Drive', 'value': 'front wheel drive'},
        {'label': 'Rear Wheel Drive', 'value': 'rear wheel drive'},
        {'label': 'ALl Wheel Drive', 'value': 'all wheel drive'},
        {'label': 'Four Wheel Drive', 'value' : 'four wheel drive'}
    ],
    value=['front wheel drive', 'rear wheel drive', 'all wheel drive', 'four wheel drive'],
    id='checklist2',
    labelStyle={'display': 'inline-block'}
    ),   
    html.Label('Vehicle Type'),
    dcc.Checklist(
    options=[
        {'label': 'Coupe', 'value': 'Coupe'},
        {'label': 'Convertible', 'value': 'Convertible'},
        {'label': 'Sedan', 'value': 'Sedan'},
        {'label': 'Wagon', 'value' : 'Wagon'},
        {'label': '4dr Hatchback', 'value' : '4dr Hatchback'},
        {'label': '2dr Hatchback', 'value' : '2dr Hatchback'},
        {'label': '4dr SUV', 'value' : '4dr SUV'},
        {'label': 'Passenger Minivan', 'value' : 'Passenger Minivan'},
        {'label': 'Cargo Minivan', 'value' : 'Cargo Minivan'},
        {'label': 'Crew Cab Pickup', 'value' : 'Crew Cab Pickup'},
        {'label': 'Regular Cab Pickup', 'value' : 'Regular Cab Pickup'},
        {'label': 'Extended Cab Pickup', 'value' : 'Extended Cab Pickup'},
        {'label': '2dr SUV', 'value' : '2dr SUV'},
        {'label': 'Cargo Van', 'value' : 'Cargo Van'},
        {'label': 'Convertible SUV', 'value' : 'Convertible SUV'},
        {'label': 'Passenger Van', 'value' : 'Passenger Van'}
    ],
    value=['Coupe', 'Convertible', 'Sedan', 'Wagon', '4dr Hatchback',
       '2dr Hatchback', '4dr SUV', 'Passenger Minivan', 'Cargo Minivan',
       'Crew Cab Pickup', 'Regular Cab Pickup', 'Extended Cab Pickup',
       '2dr SUV', 'Cargo Van', 'Convertible SUV', 'Passenger Van'],
    id='checklist3',
    labelStyle={'display': 'inline-block'}
    ),
    html.Label('MSRP Range'),
    dcc.RangeSlider(
        id='price-slider',
        min=0,
        max=1000000,
        step=5000,
        value=[0, 1000000],
        marks={0: "0",
              1000000: "1000000"}
        #allowCross=False
    ),
    
    dcc.Graph(id='graph-with-slider'),
])
@app.callback(
    Output('graph-with-slider', 'figure'),
    Input('checklist', 'value'),
    Input('checklist2', 'value'),
    Input('checklist3', 'value'),
    Input('price-slider', 'value'))
def update_figure(Selected_origin, Selected_drive, Selected_type, Selected_msrp):
    
    
    
    new_df = CarsCleaned2[(CarsCleaned2['Origin'].isin(Selected_origin))]
    new_df = new_df[(new_df['Driven_Wheels'].isin(Selected_drive))]
    new_df = new_df[(new_df['Vehicle Style'].isin(Selected_type))]    
    new_df = new_df.loc[new_df.MSRP < Selected_msrp[1]]
    new_df = new_df.loc[new_df.MSRP > Selected_msrp[0]]
                           
                           
                           
    new_df = new_df.groupby("Make").count().reset_index()
    new_df["Count"] = new_df["Origin"]
    new_df
    fig = px.histogram(new_df, x="Make", y="Count", title="Number of Cars by Make")

    fig.update_layout(transition_duration=500)

    return fig

if __name__ == '__main__':
    app.run_server(debug=False)
```
